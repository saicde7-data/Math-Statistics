## Normal Distribution
    Mean, median and mode are very close to each other
    Mean is the best measure of center. 
    Standard Deviation is the best measure of spread.

```python
n = np.random.normal(0, 1, 1000)
```

```python
f, ax = plt.subplots()

# Calculate Mean, Median & Mode
mean=np.mean(n)
median=np.median(n)

sns.distplot(n)
ax.axvline(mean, color='r', linestyle='--')
ax.axvline(median, color='g', linestyle='-')
```

## Right Skewed Distribution
When distribution is skewed to the right, mean gets pulled to the right. 

    Median is the best measure of center
    IQR is the best measure of spread - difference between the first and the third quartiles

```python
f, ax = plt.subplots()

data = scipy.stats.skewnorm(5.3, -0.1, 2.2).rvs(1000)

# Calculate Mean, Median & Mode
mean=np.mean(data)
median=np.median(data)

# estimate parameters from sample
ae, loce, scalee = scipy.stats.skewnorm.fit(data)

# Plot the PDF.
plt.hist(data, bins=100, density=True, alpha=0.6)

# Add Vertical Line for Mean, Median & Mode
ax.axvline(mean, color='r', linestyle='--')
ax.axvline(median, color='g', linestyle='-')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = scipy.stats.skewnorm.pdf(x,ae, loce, scalee)
plt.plot(x, p, 'k', linewidth=2)
```

## Left Skewed Distribution
When distribution is skewed to the left, mean gets pulled to the left

    Median is the best measure of center
    IQR is the best measure of spread - difference between the first and the third quartiles

```python
f, ax = plt.subplots()

data = scipy.stats.skewnorm(-5.3, 0.1, 2.2).rvs(1000)

mean=np.mean(data)
median=np.median(data)

# estimate parameters from sample
ae, loce, scalee = scipy.stats.skewnorm.fit(data)

# Plot the PDF.
plt.hist(data, bins=100, density=True, alpha=0.6)

# Add Vertical Line for Mean, Median & Mode
ax.axvline(mean, color='r', linestyle='--')
ax.axvline(median, color='g', linestyle='-')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = scipy.stats.skewnorm.pdf(x,ae, loce, scalee)
plt.plot(x, p, 'k', linewidth=2)
```

## Bimodal Distribution

```python
N = 100
np.random.seed(1)
f, ax = plt.subplots()

X = np.concatenate((np.random.normal(0, 1, int(0.3 * N)),
                    np.random.normal(5, 1, int(0.7 * N))))[:, np.newaxis]

mean=np.mean(X)
median=np.median(X)

ax.axvline(mean, color='r', linestyle='--')
ax.axvline(median, color='g', linestyle='-')

sns.distplot(X, bins=20)
```
