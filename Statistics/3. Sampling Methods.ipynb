## Convenience Sample

If you are an actor and want to get feedback on your acting in a movie, if you have taken feedback from your group of friends, then it is going to be considered as Convenience Sample

## Voluntary Response Sample

If you are an actor and want to get feedback on your acting in a movie, if you have approached to you followers in some Social Media Network, some of them will voluntarily come and give feedback to you. We call this kind of sample as Voluntary Response Sample

## Simple Random Sample (SRS)

If you want to start a business in a particular area and you have approached random people in that area and asked for feedback, then it is going to be called as Simple Random Sample

## Stratified Random

If you want to start a Pizza corner in a particular area, If you pick some 100 people and segregated them into two groups like Fast Food Lovers and Traditional Food Lovers, And if you take feedback from each of the groups, it is going to be considered as Stratified Random

## Multi-Stage

Multi-Stage is a combination of Simple Random Sampling and Stratified Random Sampling. Like in Stratified Random Sampling, if we divide into two or more groups and apply SRS i.e., Simple Random Sampling, then it is called as Multi-Stage Sampling as it involves multiple stages
