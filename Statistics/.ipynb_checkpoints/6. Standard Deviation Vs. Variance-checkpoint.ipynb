## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats
%matplotlib inline
```

#### Variance and Standard Deviation

Variance: The average of the squared differences from the Mean
Std: Square root of the Variance

Measure of how spread out numbers are

## Create a Dataframe

```python
df = pd.DataFrame({'Player 1': [23,46,15,0,15,22,16,5,56,49], 
              'Player 2': [46,36,49,36,0,49,53,21,25,15], 
              'Player 3': [0,0,0,245,15,0,0,2,45,23],
              'Player 4': [22,24,25,76,24,53,74,17,15,0],
              'Player 5': [45,45,53,15,12,6,15,23,30,20]}, index=['Match 1', 'Match 2', 'Match 3', 'Match 4', 'Match 5', 'Match 6', 'Match 7', 'Match 8', 'Match 9', 'Match 10'])
```

```python
df
```

## Mean of all Players

```python
df['Player 1'].mean(), df['Player 2'].mean(), df['Player 3'].mean(), df['Player 4'].mean(), df['Player 5'].mean()
```

## Score Distribution of all Players

```python
fig, ax = plt.subplots(5, 1, figsize=(10, 8), sharex=True)
sns.distplot(df['Player 1'], ax=ax[0])
sns.distplot(df['Player 2'], ax=ax[1])
sns.distplot(df['Player 3'], ax=ax[2])
sns.distplot(df['Player 4'], ax=ax[3])
sns.distplot(df['Player 5'], ax=ax[4])
plt.show()
```

```python
df['Player 2'].std(), df['Player 3'].std(), df['Player 4'].std()
```

```python
df['Player 2'].var(), df['Player 3'].var(), df['Player 4'].var()
```

```python
np.sqrt(df['Player 2'].var()), np.sqrt(df['Player 3'].var()), np.sqrt(df['Player 4'].var())
```

```python
sum = 0
for i in range(len(df)):
    sum += (df['Player 2'][i] - df['Player 2'].mean()) ** 2
```

```python
np.sqrt(sum / (len(df)-1))
```
