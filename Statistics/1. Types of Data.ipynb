# Observation

### Structured Observation

If the data has been collected by observation under certain conditions or parameters, then it is called as Structured Observation

### Unstructured Observation

When the Observer doesn't have any constraints or thoughts before observation, then it is called as Unstructured Observation

### Participant

When the Observer directly participates to collect the data, then it is called as Participant Observation

### Non-Participant

When the Observer doesn't directly participates to collect the data, then it is called as Non-Participant Observation

### Controlled

When the observation is under controlled environment like Labs, then it is called as Controlled Observation

### Uncontrolled

When the Observation occurs in Natural Environment without any Control, then it is called as Uncontrolled Observation.

# Interview

## Structured

If it contains Pre-determined Questions

## Unstructured

It if doesn't contains Pre-determined Questions

## Focussed

When the interview is been focussed on a particular experience like asking People about their experience after watching a Movie

## Clinical

When the Interview is been focussed on Life Experience rather than on a Specific Experience like Interviewing a Legendary football player like Jidan about his Life

## Group

When a group of people been interviewed at a time like Interviewing a Team of Football players in a TV Show

## Qualitative

If the Interview is on Qualitative aspects like Good or Bad, Quality of a Product, etc.,

## Quantitative

If the interview is on Quantitative aspects like a Meteorologist explaining How much rainfall it is going to be in this monsoon

## Selection

If the interview is for selecting a particular person like Audition for Selecting cast for a Movie

## Depth

If the Interview is knowing the depth of a subject like a Technical Interview of a Software Engineer to get into a an Organization like Google

## Telephonic

If the interview is going to be through Telephone rather than meeting in person
